```python
df = pd.read_csv('GDPlist.csv', encoding="latin1")
df.head(3)
```

```python
df_SA = df[(df['Continent'] == 'South America')]
df_SA.head()
```

```python
x = df_SA['Country']
y = df_SA['GDP (millions of US$)']

from matplotlib.pyplot import figure
figure(figsize=(10, 8))

plt.bar(x, y)
plt.xlabel('Country', fontsize = 12)
plt.ylabel('GDP (millions of US$)', fontsize = 12)
plt.show()
```

```python
filter_list = ['Vietnam', 'Indonesia', 'Cambodia', 'Thailand', 'Malaysia']
contains_filter = '|'.join(filter_list)
df_SEA = df[pd.notna(df.Country) & df.Country.str.contains(contains_filter)]
df_SEA
```

```python
labels = df_SEA['Country']
sizes = df_SEA['GDP (millions of US$)']
plt.pie(sizes, labels = labels)
plt.title('GDP in SEA', fontsize = 14)
plt.show()
```
